```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Loading CSV #
data = pd.read_csv('C:/Users/maxwe/Downloads/archive/Salary_Data.csv')
```

```python
# Stating the X and Y Variable for line plot #
X = data[['YearsExperience']]
y = data['Salary']
```

```python
# Split the data into training and testing sets (80% training, 20% testing) #
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Polynomial Regression features #
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
# Train polynomial regression #
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
# Predict on test data #
y_pred = model.predict(X_test_poly)
```

```python
# Evaluate the model (using Mean Squared Error for simplicity) #
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# Plot the results (train data + regression line) #
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.plot(X_train, model.predict(poly.transform(X_train)), color='red', label='Polynomial Fit')
plt.title('Salary expectations with years of experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
